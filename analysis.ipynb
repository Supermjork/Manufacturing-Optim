```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data/supply_chain_data.csv")
```

```python
data.head()

data.info()
```

```python
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns.tolist()
category_cols = data.select_dtypes(include=['object', 'category']).columns.tolist()

print(f"Numerical Columns:  {numerical_cols}")
print(f"Categorical Columns: {category_cols}")
```

```python
# Countplot for categoricals
for col in category_cols:
    plt.figure(figsize = (12, 6))
    plot = sns.countplot(x = col, data = data, palette = 'flare')
    plt.title(f"Countplot for {col}")
    plt.xticks(rotation = 45)
    plt.show()
    fig = plot.get_figure()
    fig.savefig(f"plots/Countplot_{col}.png")
```

```python
# Plotting histograms for each numerical column
for column in numerical_cols:
    plt.figure()  # Creates a new figure
    data[column].hist(bins=10, color='skyblue', edgecolor='black')  # Histogram with color
    
    # Title and axis labels
    plt.title(f'Histogram of {column}', fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    
    # Calculate mean, median, and mode
    mean_value = data[column].mean()
    median_value = data[column].median()
    mode_value = data[column].mode().values[0]  # Handling multimodal data
    
    # Add vertical lines for mean, median, and mode with distinct colors
    plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label='Mean')
    plt.axvline(median_value, color='green', linestyle='dashed', linewidth=2, label='Median')
    plt.axvline(mode_value, color='blue', linestyle='dashed', linewidth=2, label='Mode')
    
    # Add legend with larger font size
    plt.legend(fontsize=10)
    plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)  # Add grid lines for better readability
    plt.show()
```

```python
description = data[numerical_cols].describe()

print(description)
```

```python
corr = data[numerical_cols].corr()

# Generate a heatmap
plt.figure(figsize=(10, 8))  # Size of the figure
sns.heatmap(corr, annot=True, cmap='viridis', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
# Extract the correlations between independent variables and the dependent variable
correlation_with_dependent = corr['Revenue generated']

# Sort the correlations in descending order
sorted_correlations = correlation_with_dependent.abs().sort_values(ascending=False)

# Display the attributes in descending order of correlation with the dependent variable
for attribute in sorted_correlations.index:
    correlation_percent = sorted_correlations[attribute] * 100
    print(f'{attribute} and dependent variable = {correlation_percent:.2f}%')
```
